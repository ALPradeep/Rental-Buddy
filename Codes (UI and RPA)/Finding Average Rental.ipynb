```python
file_list = os.listdir()
```

```python
mrt_stations = []
ave_prices = []

for file_name in file_list:
    data = pd.read_csv(file_name)
    data.dropna(inplace = True)
    raw_prices = data['Price']
    
    if len(raw_prices) == 0:
        ave_prices.append("No data")
        print(f"No data for calculation of {file_name.replace('.csv','')} prices")
        print()
        
    else:
        total = 0
        for raw_price in raw_prices:
    #         print(raw_price)
            price = raw_price[raw_price.find("$")+1:raw_price.find("/mo")]
            price = price.replace(",","")
            total += int(price)

        print(f"Sum of Prices for {file_name.replace('.csv','')}: ${total}")

        ave_price = total / len(data['Price'])
        print(f"Average Rental Price for {file_name.replace('.csv','')}: ${math.ceil(ave_price)}")
        print()
        ave_prices.append(math.ceil(ave_price))
    mrt_stations.append(file_name.replace('.csv',''))
```

```python
print(ave_prices)
print(mrt_stations)
```

```python
average_rental_list = pd.DataFrame({"MRT Station": mrt_stations,
                                   "Average Rental Price":ave_prices})
average_rental_list.to_excel("Average Rental Price List.xlsx",index=False)
```
